```python
train, val, test = np.split(df1.sample(frac=1), [int(0.6*len(df1)), int(0.8*len(df1))])
```

common function

```python
def get_xy(dataframe, y_label, x_labels=None):
  dataframe = copy.deepcopy(dataframe)
  if x_labels is None:
    X = dataframe[[c for c in dataframe.columns if c!=y_label]].values
  else:
    if len(x_labels) == 1:
      X = dataframe[x_labels[0]].values.reshape(-1, 1)
    else:
      X = dataframe[x_labels].values

  y = dataframe[y_label].values.reshape(-1, 1)
  data = np.hstack((X, y))

  return data, X, y
```

```python
_, X_train_date, y_train_date = get_xy(train, "Close", x_labels=["Date"])
_, X_val_date, y_val_date = get_xy(val, "Close", x_labels=["Date"])
_, X_test_date, y_test_date = get_xy(test, "Close", x_labels=["Date"])
```
